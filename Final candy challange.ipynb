```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import statsmodels.api as sm
import scipy.stats as stats

# Load the dataset
try:
    df = pd.read_csv('D:\\PBI Projects\\Halloween+Candy+Rankings\\candy-data.csv')
except Exception as e:
    print(f"Error loading data: {e}")
```

```
# Perform linear regression for each flavor type
flavors = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 
           'crispedricewafer', 'hard', 'bar', 'pluribus']

# Perform linear regression and scatter plot analysis for each flavor
for flavor in flavors:
    # Prepare the data
    flavor_data = df[df[flavor].notna()]  # Drop rows where the flavor is NaN
    X = flavor_data['pricepercent']  # Independent variable
    y = flavor_data['winpercent']     # Dependent variable
    # Add a constant to the independent variable
    X = sm.add_constant(X)
    # Fit the linear regression model
    model = sm.OLS(y, X).fit()
    
    # Print the regression summary
    print(f'Linear Regression Summary for {flavor.capitalize()} Flavor:')
    print(model.summary())
    print('\n')

    # Create scatter plot with regression line
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=flavor_data, x='pricepercent', y='winpercent', 
                    color='blue', marker='o', s=50, label='Data Points', alpha=0.7)

    # Predict y values using the model
    predictions = model.predict(X)

    # Plot the regression line
    plt.plot(flavor_data['pricepercent'], predictions, color='red', linewidth=2, label='Regression Line')

    # Customize the plot
    plt.title(f'{flavor.capitalize()} Flavor Linear Regression', fontsize=16)
    plt.xlabel('Price Percent', fontsize=14)
    plt.ylabel('Win Percent', fontsize=14)
    plt.legend()

    # Show the plot
    plt.show()
```

```
# Store slopes (tangents) in a dictionary
slope_values = {}

# Perform linear regression for each flavor
for flavor in flavors:
    # Filter the data for the specific flavor (where flavor column is 1)
    flavor_data = df[df[flavor] == 1] 
    if not flavor_data.empty:
        # Define independent and dependent variables
        X = flavor_data[['pricepercent']]
        y = flavor_data['winpercent']
        # Add a constant (intercept) to the independent variable
        X = sm.add_constant(X)
        # Fit the linear regression model
        model = sm.OLS(y, X).fit()
        # Extract the slope (tangent) for pricepercent
        slope = model.params['pricepercent']
        # Store the slope in the dictionary with the flavor as the key
        slope_values[flavor] = slope

# Set a style
sns.set(style="white")

# Create a joint grid with KDE and scatter plot
g = sns.JointGrid(data=df, x='pricepercent', y='winpercent', space=0)

# Add scatter plot
g.plot(sns.scatterplot, sns.kdeplot)

# Add KDE on top of scatter plot with contours
g.plot_joint(sns.kdeplot, cmap="Purples", fill=True)

# Add KDE for the margins without histograms
g.plot_marginals(sns.kdeplot, fill=True, color="purple")

# Add scatter points with cross markers, change color to black
g.ax_joint.scatter(df['pricepercent'], df['winpercent'], color="black", marker='+', s=50)

# Customize labels and title
g.set_axis_labels('Price Percent', 'Win Percent')

plt.tight_layout()
plt.show()
```

```
# Create a DataFrame from the slope values
slope_df = pd.DataFrame(list(slope_values.items()), columns=['Flavor', 'Tangent'])

# Bar chart to visualize the tangent values for each flavor
plt.figure(figsize=(10, 6))
sns.barplot(x='Tangent', y='Flavor', data=slope_df, palette='coolwarm')
plt.title('Tangent (Slope) Values for Candy Flavors')
plt.xlabel('Tangent (Slope)')
plt.ylabel('Flavor')
plt.legend()
plt.grid()
plt.show()
```

```
# Perform t-tests for each flavor/type
flavors_types = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 
                 'crispedricewafer', 'hard', 'bar', 'pluribus']

# Initialize a dictionary to store t-test results for each flavor/type
t_test_results = {}

for flavor in flavors_types:
    t_test_results[flavor] = stats.ttest_ind(df[df['winpercent'] < 0.5][flavor], 
                                              df[df['winpercent'] >= 0.5][flavor])
    print(f"{flavor.capitalize()} t-test: {t_test_results[flavor]}")

# Display the t-test results in a more readable format
t_test_results_df = pd.DataFrame(t_test_results, index=['statistic', 'pvalue']).T
t_test_results_df['significant'] = t_test_results_df['pvalue'] < 0.05  # Mark if p-value is significant
print(t_test_results_df)
```

```
# Calculate mean prevalence of each flavor/type in winners and losers
winners_means = df[df['winpercent'] >= 0.5][flavors_types].mean()
losers_means = df[df['winpercent'] < 0.5][flavors_types].mean()
print(winners_means, losers_means)
```

```
# Create a bar plot to compare candy flavors between winners and losers
bar_width = 0.35
index = np.arange(len(flavors_types))

fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(index, winners_means, bar_width, label='Winners', color='green')
bar2 = ax.bar(index + bar_width, losers_means, bar_width, label='Losers', color='red')

# Adding labels and title
ax.set_xlabel('Candy Type/Flavor')
ax.set_ylabel('Average Prevalence')
ax.set_title('Comparison of Candy Types/Flavors Between Winners and Losers')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(flavors_types, rotation=45)
ax.legend()

plt.tight_layout()
plt.show()
```

```
# PCA Analysis
features = df.columns[1:13]  # Assuming the first column is 'competitorname' and the rest are features
x = df.loc[:, features].values
x = StandardScaler().fit_transform(x)

pca = PCA(n_components=4)
df_pca = pca.fit_transform(x)

# Explained variance
print("Explained Variance Ratios:", pca.explained_variance_ratio_)

# Scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.show()

# Plot principal components
plt.figure(figsize=(12, 8))
plt.scatter(df_pca[:, 0], df_pca[:, 1], s=50)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Results')
plt.show()
```

```
# Clustering using k-means after PCA
cluster_qty = 3
kmeans = KMeans(n_clusters=cluster_qty)
df['cluster'] = kmeans.fit_predict(df_pca[:, :4])

# Define cluster colors
colors = ['#af231c', '#eeb007', '#7b4931']
df['cluster_color'] = df['cluster'].map({0: colors[0], 1: colors[1], 2: colors[2]})

# Plot clusters after PCA
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=df['cluster_color'], s=50, alpha=0.7)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title("Halloween Candy Clusters")
plt.show()
```

```
# Clustering on sweetness index vs winpercent
df['sweetness_index'] = df['sugarpercent'] * df['chocolate']
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster_sweetness'] = kmeans.fit_predict(df.drop(['competitorname', 'winpercent', 'cluster_color'], axis=1))

# Visualize clusters with sweetness vs winpercent
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sweetness_index', y='winpercent', hue='cluster_sweetness', palette='Set2', data=df)
plt.title('Clustering of Candies Based on Sweetness and Popularity')
plt.show()
```

```
# Create a 'high_win' column based on winpercent
df['high_win'] = (df['winpercent'] > 60).astype(int)  # 1 for high win, 0 for not

# Box plot for winpercent distribution
plt.figure(figsize=(10, 6))
sns.boxplot(x='high_win', y='winpercent', data=df)  # Use df directly
plt.title('Win Percentage Distribution by High Win')
plt.xlabel('High Win (1=Yes, 0=No)')
plt.ylabel('Win Percentage')
plt.show()
```

```
# Violin plot for sugar percentage distribution
plt.figure(figsize=(10, 6))
sns.violinplot(x='high_win', y='sugarpercent', data=df)  # Use df directly
plt.title('Sugar Percentage Distribution by High Win')
plt.xlabel('High Win (1=Yes, 0=No)')
plt.ylabel('Sugar Percentage')
plt.show()
```

```
# Feature Importance using Gradient Boosting
# Select only the specified features for feature importance analysis
features = [
    'sugarpercent', 
    'chocolate', 
    'fruity', 
    'caramel', 
    'peanutyalmondy', 
    'nougat', 
    'crispedricewafer', 
    'hard', 
    'bar', 
    'pluribus', 
    'pricepercent'
]

# Prepare the feature set with the specified columns
X = df[features]  # Only include the specified features
y = df['winpercent']  # Target variable

# Initialize and fit the Gradient Boosting Regressor with Grid Search for hyperparameter tuning
gb_model = GradientBoostingRegressor()

# Define parameters for tuning
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5],
}

# Perform Grid Search
grid_search = GridSearchCV(gb_model, param_grid, cv=5, scoring='r2')
grid_search.fit(X, y)

# Best model from Grid Search
best_gb_model = grid_search.best_estimator_

# Get feature importance
feature_importances = pd.Series(best_gb_model.feature_importances_, index=X.columns).sort_values(ascending=False)

# Visualization of Feature Importance
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances, y=feature_importances.index)
plt.title('Feature Importance from Gradient Boosting (Selected Features Only)')
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.show()
```

```
# Data Preprocessing: Label encoding for categorical features
le = LabelEncoder()
for feature in ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 
                'crispedricewafer', 'hard', 'bar', 'pluribus']:
    df[feature] = le.fit_transform(df[feature])

# Weighted Scoring System for ranking candies
scaler = MinMaxScaler()
df[['sugarpercent_scaled', 'pricepercent_scaled', 'winpercent_scaled']] = scaler.fit_transform(
    df[['sugarpercent', 'pricepercent', 'winpercent']])

# Calculate a flavor score as the sum of all flavor indicators for each candy
df['flavor_score'] = df[['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 
                          'nougat', 'crispedricewafer', 'hard', 'bar', 
                          'pluribus']].sum(axis=1)
df['final_score'] = (0.3 * df['sugarpercent_scaled'] + 
                     0.2 * df['pricepercent_scaled'] + 
                     0.5 * df['winpercent_scaled'] +
                     0.1 * df['flavor_score'])  # Including flavor score

# Top candies
top_candies = df[['competitorname', 'final_score']].sort_values(by='final_score', ascending=False).head(10)
print("Top Candies Based on Final Score:")
print(top_candies)

# Descriptive statistics of the dataset
desc_stats = df.describe()
print("Descriptive Statistics of the Dataset:")
print(desc_stats)

# First, check the DataFrame types to identify non-numeric columns
print("DataFrame Types:")
print(df.dtypes)

# Selecting only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=[np.number])

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numeric Features')
plt.show()


```
